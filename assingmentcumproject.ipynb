```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/drive/MyDrive/Airbnb NYC 2019.csv')
display(df.head())
```

```python
display(df.columns)
```

```python
df.info()
```

cheking if the columns contain any null values

```python
df.isnull().sum()
```

in the position of name ,host_name , reviews_per_month,last_review filling the following Values Unknown,0,No Review etc

```python
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['last_review'] = df['last_review'].fillna('No Review')
```

confirming the chage

```python
df.isnull().sum()
```

```python
df['neighbourhood'].value_counts().head(10)
```

```python
df['room_type'].value_counts()
```

Top 10 Neighbourhoods by Reviews

```python
plt.figure(figsize=(10,6))
top10_reviews = df.groupby('neighbourhood')['number_of_reviews'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top10_reviews.values, y=top10_reviews.index,color='red')
plt.title("Top 10 Neighbourhoods by Reviews")
plt.xlabel("Average Number of Reviews")
plt.ylabel("Neighbourhood")
plt.show()
#Neighbourhoods with higher average reviews are likely preferred by guests
```

# Which neighbourhoods or groups have the highest average price?

```python
# Average price by neighbourhood_group
avg_price_group = df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
print(avg_price_group)
```

```python
sns.barplot(x=avg_price_group.index, y=avg_price_group.values)
plt.title('Average Price by Neighbourhood Group')
plt.ylabel('Average Price')
plt.show()
```

```python
# Average price by neighbourhood (Top 10 expensive)
avg_price_neigh = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10)
print(avg_price_neigh)
avg_price_neigh.plot(kind='bar', figsize=(10,6), title='Top 10 Expensive Neighbourhoods',color='purple')
plt.ylabel('Average Price')
plt.show()
#This will show which regions are expensive
```

# Whats the distribution of prices

```python
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.show()

sns.boxplot(x=df['price'])
plt.title('Price Boxplot )')
plt.show()
```

# How does room_type affect the average price?

```python
# Average price by room_type
avg_price_room = df.groupby('room_type')['price'].mean()
print(avg_price_room)
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price Distribution by Room Type')
plt.show()
```

#Which neighbourhoods get the most reviews?

```python
# Total reviews per neighbourhood
reviews_neigh = df.groupby('neighbourhood')['number_of_reviews'].sum().sort_values(ascending=False).head(10)
print(reviews_neigh)

reviews_neigh.plot(kind='bar', figsize=(10,6), title='Top 10 Neighbourhoods by Number of Reviews')
plt.ylabel('Total Reviews')
plt.show()
#Helps see how "Entire home/apt" compares to "Private room"
```

# How does availability_365 relate to price or reviews_per_month

```python
#  availability vs price
sns.scatterplot(x='availability_365', y='price', data=df,color='green')
plt.title('Availability vs Price')
plt.show()

#  availability vs reviews_per_month
sns.scatterplot(x='availability_365', y='reviews_per_month', data=df)
plt.title('Availability vs Reviews per Month')
plt.show()
#Shows if more available listings get more reviews or have different prices.
```

# Which hosts have the most listings or reviews

```python
# Hosts with most listings
top_hosts = df['host_name'].value_counts().head(10)
print(top_hosts)

top_hosts.plot(kind='bar', figsize=(10,6), title='Top 10 Hosts by Number of Listings')
plt.ylabel('Number of Listings')
plt.show()

# Hosts with most reviews (sum of reviews per host)
top_hosts_reviews = df.groupby('host_name')['number_of_reviews'].sum().sort_values(ascending=False).head(10)
print(top_hosts_reviews)

top_hosts_reviews.plot(kind='bar', figsize=(10,6), title='Top 10 Hosts by Total Reviews')
plt.ylabel('Number of Reviews')
plt.show()
```

# Where are the most expensive listings located (using latitude & longitude)

```python
# Top 100 expensive listings
top_expensive = df.sort_values('price', ascending=False).head(100)
plt.figure(figsize=(10,8))
plt.scatter(top_expensive['longitude'], top_expensive['latitude'], c=top_expensive['price'], cmap='Reds', s=50)
plt.colorbar(label='Price')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Top 100 Expensive Listings Location')
plt.show()
```

```python
top_neigh = df.groupby('neighbourhood')['number_of_reviews'].sum().sort_values(ascending=False).head(10)
neigh_list = list(top_neigh.index)

df_top = df[df['neighbourhood'].isin(neigh_list)]
listings_count = df_top['neighbourhood'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=listings_count.index, y=listings_count.values, palette='Blues_r')
plt.title('Number of Listings by Neighbourhood')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df_top, x='neighbourhood', hue='room_type')
plt.title('Room Type Distribution by Neighbourhood')
plt.xticks(rotation=45)
plt.ylabel('Number of Listings')
plt.show()
```

```python
avg_reviews_month = df_top.groupby('neighbourhood')['reviews_per_month'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_reviews_month.index, y=avg_reviews_month.values, palette='Greens_r')
plt.title('Average Reviews per Month by Neighbourhood')
plt.ylabel('Reviews per Month')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='neighbourhood', y='availability_365', data=df_top)
plt.title('Listing Availability by Neighbourhood')
plt.xticks(rotation=45)
plt.ylabel('Availability (Days)')
plt.show()
```
